```python
import matplotlib.pyplot as plt 
import matplotlib.lines as lines
import numpy as np 
```

```python
#  oscillating pendulum animation for no air resistance
plt.ion()     

dt = 0.01
t_max = 100

fig, ax = plt.subplots()

#initial  objects
line_obj = None
bob_obj = None

ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')

def animate(theta):
    global line_obj, bob_obj

    # remove previous frame
    if line_obj is not None:
        line_obj.remove()
    if bob_obj is not None:
        bob_obj.remove()

    x = np.sin(theta)
    y = -np.cos(theta)

    # redraw
    line_obj = ax.add_line(lines.Line2D([0, x], [0, y], lw=3, color='r'))
    bob_obj = ax.scatter(x, y, s=80, c='r')

    plt.pause(0.001)   


def phase_trajectory(theta_0, dot_theta_0):
    theta = theta_0
    dot_theta = dot_theta_0
    t = 0
    steps=0

    while t < t_max:
        if steps%10==0:
            animate(theta)

       
        dot_theta = dot_theta - dt*np.sin(theta)
        theta = theta + dt*dot_theta

        t += dt
        steps+=1


phase_trajectory(np.pi/6, 0)
```
