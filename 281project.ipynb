```python
import matplotlib.pyplot as plt 
import matplotlib.lines as lines
import numpy as np 
```

```python


dt = 0.001
t_max = 100
def phase_trajectory(theta_0, dot_theta_0):
    theta = theta_0
    dot_theta = dot_theta_0
    t = 0
    thetalst=[theta]
    times=[t]
    dot_thetalst=[dot_theta]
    intialEnergy=(dot_theta**2/2)-np.cos(theta)
    Energylst=[intialEnergy]
   
    while t < t_max:
        

        dot_theta = dot_theta - dt*np.sin(theta)
        theta = theta + dt*dot_theta
    
        thetalst.append(theta)
        times.append(t)
        dot_thetalst.append(dot_theta)
        Energylst.append((dot_theta**2/2)-np.cos(theta))
        t+=dt
    return thetalst, times ,dot_thetalst,Energylst
intial_theta_values=np.linspace(-2*np.pi,2*np.pi,10)
for i in intial_theta_values:

    theta,time,dotttheta,Energy=phase_trajectory(i,0)
    if abs(Energy[0])>1:
        Color='red'
    else:
        Color="blue"
    Theta=[]
    for i in theta:
        Theta.append(i/np.pi)
    plt.scatter(Theta,dotttheta,color=Color,s=1)
intial_theta_values=np.linspace(-1*np.pi,1*np.pi,10)
for i in intial_theta_values:
  
    theta,time,dotttheta,Energy=phase_trajectory(i,1)
    if abs(Energy[0])>1:
        Color='red'
    else:
        Color="blue"
    Theta=[]
    for i in theta:
        Theta.append(i/np.pi)
        
    plt.scatter(Theta,dotttheta,color=Color,s=1)
intial_theta_values=np.linspace(-2*np.pi,2*np.pi,10)
for i in intial_theta_values:
    
    theta,time,dotttheta,Energy=phase_trajectory(i,-1)
    if abs(Energy[0])>1:
        Color='red'
    else:
        Color="blue"
    Theta=[]
    for i in theta:
        Theta.append(i/np.pi)
    plt.scatter(Theta,dotttheta,color=Color,s=1)
plt.xlim(-1.1,1.1)
plt.xlabel("theta/pi")
plt.ylabel("dottheta")

plt.scatter(0, 0, color='black', s=40)  
plt.scatter(1, 0, color='black', s=40)   
plt.scatter(-1, 0, color='black', s=40)  
plt.show()


theta,time,dotttheta,Energy=phase_trajectory(np.pi/4,0)

plt.scatter(time,Energy,s=1)  
plt.ylim(-.5,-1)  
plt.xlabel("time")
plt.ylabel("Energy")
plt.show()

theta,time,dotttheta,Energy=phase_trajectory(np.pi/2,3)

plt.scatter(time,Energy,s=1)  
plt.ylim(4.3,4.6)
plt.xlabel("time")
plt.ylabel("Energy")
plt.show()
theta,time,dotttheta,Energy=phase_trajectory(np.pi/1.2, 1)
Theta=[]
for i in theta:
    Theta.append(i/np.pi)
plt.scatter(Theta,dotttheta,s=1)
plt.xlim(0,4*np.pi)
plt.xlabel(r'$\theta/\pi$')
plt.ylabel(r'$\dot{\theta}$')
plt.title(f'Phase portrait for intial theta {Theta[0]} pi and intial angular velocity {dotttheta[0]}')
plt.grid(True)
plt.show()
  

theta,time,dotttheta,Energy=phase_trajectory(np.pi/4, 0)
Theta=[]
for i in theta:
    Theta.append(i/np.pi)
plt.scatter(Theta,dotttheta,s=1)
plt.xlabel(r'$\theta/\pi$')
plt.ylabel(r'$\dot{\theta}$')
plt.title(f'Phase portrait for intial theta {Theta[0]} pi and intial angular velocity {dotttheta[0]}')
plt.grid(True)
plt.show()
```
